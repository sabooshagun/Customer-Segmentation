## Segmentation on the basis of Users (Top 10)

```python
User=data['UserId'].unique()
User
```

```python
User_collection = {} 

for i in User:
    new_datauser = data.loc[data["UserId"]==i]
    User_collection[i] = new_datauser
```

```python
usertotal = []

for i in User:
    usertotal.append(User_collection[i].TotalSale.sum()) 
```

```python
usertotalsort=usertotal
```

```python
len(User)
```

```python
def partition(arr,low,high): 
    i = ( low-1 )         # index of smaller element 
    pivot = arr[high]     # pivot 
  
    for j in range(low , high): 
  
        # If current element is smaller than or 
        # equal to pivot 
        if   arr[j] <= pivot: 
          
            # increment index of smaller element 
            i = i+1 
            arr[i],arr[j] = arr[j],arr[i] 
  
    arr[i+1],arr[high] = arr[high],arr[i+1] 
    return ( i+1 ) 
  
# The main function that implements QuickSort 
# arr[] --> Array to be sorted, 
# low  --> Starting index, 
# high  --> Ending index 
  
# Function to do Quick sort 
def quickSort(arr,low,high): 
    if low < high: 
  
        # pi is partitioning index, arr[p] is now 
        # at right place 
        pi = partition(arr,low,high) 
  
        # Separately sort elements before 
        # partition and after partition 
        quickSort(arr, low, pi-1) 
        quickSort(arr, pi+1, high) 
```

```python

n = len(usertotalsort) 
quickSort(usertotalsort,0,n-1) 
```

```python
top10=[]
for i in reversed(range (n-10,n)):
    top10.append(usertotalsort[i])

                
```

```python
top10
```

```python
top10usr=[]
for i in range (len(top10)):
    for j in range (len(usertotal)):
               
        if top10[i]==usertotal[j]:
            top10usr.append(User[j])
```

```python
top10usr
```

```python
df=pd.DataFrame({'top10usr':top10usr,'top10':top10})
df
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])

ax.bar(df['top10usr'].astype(str),df['top10'])
plt.xticks(rotation=90)
plt.xlabel("UserId")
plt.ylabel("SALE")
plt.title("Segmenation on the basis of Country")
plt.show()
```
